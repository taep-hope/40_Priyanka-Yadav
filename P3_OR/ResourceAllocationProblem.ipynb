Resource Allocation Problem

```
from scipy.optimize import linprog
```

```
obj = [-20, -12, -40, -25]
```

```
lhs_ineq = [[1, 1, 1, 1],
            [3, 2, 1, 0],
            [0, 1, 2, 0]]
```

```
rhs_ineq = [50, 100, 90]
```

```
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq, method="revised simplex")
```
